```python
from Bio import AlignIO, Phylo
from Bio.Phylo.TreeConstruction import DistanceCalculator, DistanceTreeConstructor
```

```python
aln_GH42 = AlignIO.read("MSA_GH42_Copy.sth", "stockholm")

print(f"Alignment: \n {aln_GH42} \n ####################")
```

```python
calculator = DistanceCalculator('blosum62')
dm = calculator.get_distance(aln_GH42)
print(f"Distance matrix: \n{dm} \n ##########################")
```

```python
constructor_GH42 = DistanceTreeConstructor()

upgmatree_GH42 = constructor_GH42.upgma(dm)

print(f"UPGMA tree: \n{upgmatree_GH42} \n ###################")

#salvar os arquivos 

Phylo.write(upgmatree_GH42, "upgma_gh42_copy.nwk", "newick")
```

```python
njtree_GH42 = constructor_GH42.nj(dm)

print(f"NJ tree: \n{njtree_GH42} \n #########################")

#Salvar arquivo

Phylo.write(njtree_GH42, "nj.nwk_gh42_copy", "newick")
```
